```python
import matplotlib.pyplot as plt
import math
%matplotlib inline
plt.style.use('seaborn')
```

```python
x_axis = []
y_axis = []
for i in range(158):
    x_axis.append(i)
    value_y = (i/(4*(math.pi**2 *(5*10**-9)*(1.6*10**-19))))**0.5
    y_axis.append(value_y)

plt.ylabel('Freq(Hz)')
plt.xlabel('Volt(V)')
plt.title('Question 1E')
plt.plot(x_axis,y_axis, color='green', linewidth=3)
plt.show()
```
